```python
# Handling missing values
```

```python
df.isnull().sum()
```

```python
# look at the missing records once

df[df.Released_Year.isnull()]
df[df.Meta_score.isnull()]
df[df.Gross.isnull()]
```

```python
# filling missing values
```

```python
#because the most movies is "U" :General Viewing for all ages 
df['Certificate'].fillna('U',inplace=True)

df['Meta_score'].fillna(df['Meta_score'].mean(),inplace=True)
```

```python
#convert type to int to deal with it 
df["Gross"] = df["Gross"].str.replace(",","")
df["Gross"] = df["Gross"].replace(np.nan, 0)
df["Gross"] = df["Gross"].astype(float)

# fill null data with the avg gross
df["Gross"] = df["Gross"].replace(0,df['Gross'].mean())
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["Runtime"] = df["Runtime"].str.extract('([^\s]+)')
df["Runtime"] = df["Runtime"][~(df["Runtime"] == "min")]
df["Runtime"] = df["Runtime"].astype(int)
```

```python
#drop row contain 'PG' to convert the column to int datatype
df = df.drop(labels=966, axis=0)
```

```python
df['Released_Year']=df['Released_Year'].astype(int)
```

## Percentage of Genre

```python
gen = []
for x in df.Genre:
    for y in x.split(','):
        gen.append(y.strip())


genre=((pd.Series(gen).value_counts()/len(gen) )*100).round(2).head(12)
plt.figure(figsize=(16,9))
plt.pie(x=genre.values,
        labels=genre.index,
        autopct='%.2f%%')
plt.title('Percentage of Genre',weight='bold');
```

##### Drama movies have been most successful with overall 32.37 % in top 999 series dataset

## Directors with highest number of votes

```python
top_10=df.nlargest(10,'No_of_Votes')[['Director','No_of_Votes','IMDB_Rating','Genre']]
top_10
```

```python
plt.figure(figsize=(16,7))
sns.barplot(y=top_10.No_of_Votes,x='Director',data=top_10,ci=0,palette='Oranges',hue='IMDB_Rating',dodge=False)
plt.title('Director with highest number of votes',weight='bold')
plt.xticks(rotation=15)
plt.ylabel('Votes (in millions)');
```
